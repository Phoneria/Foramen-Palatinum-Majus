```python
def move_file(source_dir,destination_dir):
    # Get all folders in the source directory
    for folder_name in os.listdir(source_dir):
        folder_path = os.path.join(source_dir, folder_name)
        destination_folder_path = os.path.join(destination_dir, folder_name)
        
        # Check if it's a folder
        if os.path.isdir(folder_path):
            # Check if the destination folder already exists
            if os.path.exists(destination_folder_path):
                print(f"Skipping: {folder_name} (already exists in destination)")
            else:
                # Move the folder to the destination directory
                shutil.move(folder_path, destination_dir)
                print(f"Moved: {folder_name}")

    print("Folders have been moved, skipping existing ones.")
```

```python
# Define source and destination paths
destination_dir = tomo_data_path

# Move 2017 tomo data
move_file(tomo_2017,destination_dir)

# Move 2018 tomo data
for folder_18 in os.listdir(tomo_2018):
    move_file(os.path.join(tomo_2018,folder_18),destination_dir)

# Move 2019 tomo data
move_file(tomo_2019,destination_dir)

# Move april tomo data
move_file(tomo_april,destination_dir)


shutil.rmtree(tomo_2017)
shutil.rmtree(tomo_2018)
shutil.rmtree(tomo_2019)
shutil.rmtree(tomo_april)
```

```python
label_list = os.listdir(label_data_path)
tomos_list = os.listdir(tomo_data_path)
```

```python
patients_tomos = sorted([name.split("\\")[-1].lower() for name in tomos_list])
patients_labels = sorted([name.split(".")[0].lower() for name in label_list])
```

```python
print("Number of Tomos : ",len(patients_tomos), "\nNumber of Labels : ", len(patients_labels))
```

```python
# Dictionary to store matched and unmatched names
matched_names = []
unmatched_names = []

counter = 0
# Match names using fuzzy matching
for tomo in patients_tomos:
    print(len(patients_tomos) - counter)
    counter+=1
    
    label, score = process.extractOne(tomo, patients_labels)
    if score > 80:  # Set a threshold for matching
        matched_names.append((tomo, label))  # Store original and matched name
    else:
        unmatched_names.append(tomo)

# Output the results
print("Matched Names:")
for original, match in matched_names:
    print(f"Original: {original} -> Matched: {match}")

print("\nUnmatched Names:")
print(unmatched_names)
```

```python
# Dictionary to hold normalized names and their original names
patience_names = {}

# Populate the reverse mapping
for original, matched in matched_names:
    if matched not in patience_names:
        patience_names[matched] = []  # Initialize a list for new normalized value
    patience_names[matched].append(original)  # Append the original name to the list
```

```python
# Filter to find names with the same normalized value
same_value_keys = {patience_names: names for patience_names, names in patience_names.items() if len(names) > 1}


# Output the results
print("Names with the same normalized version:")
for normalized, names in same_value_keys.items():
    print(f"Normalized: '{normalized}' -> Original Names: {names}")
```

```python
import os
from dicom2jpg import dicom2png  # Ensure that dicom2jpg is installed and imported

# Define the main directory and the export location
main_dir = r'C:\Users\Nasip-PC\Desktop\Dosyalar\Kodlar\TEZ 2024\tomolar'
export_location = r'C:\Users\Nasip-PC\Desktop\Dosyalar\Kodlar\TEZ 2024\tomo_png'

# Create the export directory if it doesn't exist
if not os.path.exists(export_location):
    os.makedirs(export_location)

# Iterate through each subdirectory in the main directory
for foldername in os.listdir(main_dir):
    folder_path = os.path.join(main_dir, foldername)

    # Check if it's a directory
    if os.path.isdir(folder_path):
        # Create a corresponding directory in the export location
        new_folder_path = os.path.join(export_location, foldername)
        if not os.path.exists(new_folder_path):
            os.makedirs(new_folder_path)

        # Iterate through each file in the folder
        for filename in os.listdir(folder_path):
            if filename.endswith('.dcm'):
                dicom_file_path = os.path.join(folder_path, filename)
                png_file_name = os.path.splitext(filename)[0] + '.png'
                png_file_path = os.path.join(new_folder_path, png_file_name)

                # Convert DICOM to PNG
                try:
                    dicom2png(dicom_file_path, png_file_path)
                    print(f'Converted {dicom_file_path} to {png_file_path}')
                except Exception as e:
                    print(f'Failed to convert {dicom_file_path}: {e}')
```
