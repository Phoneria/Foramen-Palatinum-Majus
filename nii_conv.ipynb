```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

def convert_nifti_to_png(nifti_dir, output_dir):
    counter = 0
    # Iterate over each NIfTI file in the directory
    for filename in os.listdir(nifti_dir):
        if filename.endswith('.nii') or filename.endswith('.nii.gz'):
            # Load the NIfTI file
            nifti_path = os.path.join(nifti_dir, filename)
            img = nib.load(nifti_path)
            img_data = img.get_fdata()

            # Create a patient-specific output directory
            patient_name = os.path.splitext(filename)[0]  # Use filename without extension as patient name
            patient_dir = os.path.join(output_dir, patient_name)
            os.makedirs(patient_dir, exist_ok=True)

            # Create directories for different orientations
            axial_dir = os.path.join(patient_dir, 'axial_slices')
            sagittal_dir = os.path.join(patient_dir, 'sagittal_slices')
            coronal_dir = os.path.join(patient_dir, 'coronal_slices')

            os.makedirs(axial_dir, exist_ok=True)
            os.makedirs(sagittal_dir, exist_ok=True)
            os.makedirs(coronal_dir, exist_ok=True)

            # Save Axial slices
            for i in range(img_data.shape[2]):  # Axial slices
                slice_data = img_data[:, :, i]
                slice_data_normalized = normalize_slice(slice_data)
                plt.imsave(os.path.join(axial_dir, f"{i:04d}.png"), slice_data_normalized, cmap='gray')

            # Save Sagittal slices
            for i in range(img_data.shape[0]):  # Sagittal slices
                slice_data = img_data[i, :, :]
                slice_data_normalized = normalize_slice(slice_data)
                plt.imsave(os.path.join(sagittal_dir, f"{i:04d}.png"), slice_data_normalized, cmap='gray')

            # Save Coronal slices
            for i in range(img_data.shape[1]):  # Coronal slices
                slice_data = img_data[:, i, :]
                slice_data_normalized = normalize_slice(slice_data)
                plt.imsave(os.path.join(coronal_dir, f"{i:04d}.png"), slice_data_normalized, cmap='gray')
        print(len(os.listdir(nifti_dir)) - counter, " file left")
        counter += 1

def normalize_slice(slice_data):
    """Normalize the slice data to range 0-255."""
    slice_data_normalized = ((slice_data - np.min(slice_data)) / (np.max(slice_data) - np.min(slice_data)) * 255).astype(np.uint8)
    return slice_data_normalized
```

```python

# Define your input and output directories
nifti_directory = r'C:\Users\Nasip-PC\Desktop\Dosyalar\Kodlar\TEZ 2024\etiketler_unzipped'  # Change this to your NIfTI directory
output_directory = r'C:\Users\Nasip-PC\Desktop\Dosyalar\Kodlar\TEZ 2024\etiketler_png'  # Change this to your desired output directory

# Convert NIfTI files to PNG
convert_nifti_to_png(nifti_directory, output_directory)

print("Conversion complete.")
```
